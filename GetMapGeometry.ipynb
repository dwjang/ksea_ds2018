```python
import requests, pprint

def get_geometry(app_id, app_code):

    url = "https://pde.api.here.com/1/search/bbox.json"

    params = {
        "app_id": app_id,
        "app_code": app_code,
        "layer_id": "LINK_FC2",
        "key_attribute": "LINK_ID",
        "bbox": "41.904,-87.682;41.853,-87.619",
    }

    result = requests.get(url=url, params=params)

    print("status code:", result.status_code)

    if result.status_code == 200:
        content = result.json()
        pprint.pprint(content)
        
    links = []
    for link in content["geometries"]:
        linkid = link["attributes"]["LINK_ID"]
        linklength = link["attributes"]["LINK_LENGTH"]
        startnodeid = link["attributes"]["REF_NODE_ID"]
        endnodeid = link["attributes"]["NONREF_NODE_ID"]
        geoms = link["geometry"].replace("MULTILINESTRING((","").replace("))","").split(",")
        shapepoints = [x.split(" ") for x in geoms]
        links.append([linkid, linklength, startnodeid, endnodeid,shapepoints])

    return links
```

```python
app_id = "gtswh2sa822DHthznV1X"
app_code = "zd2FnIRxq25-BNHJSqgjtg"
links = get_geometry(app_id, app_code)
```
